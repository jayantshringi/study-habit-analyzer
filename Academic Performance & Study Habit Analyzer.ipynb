#  Academic Performance & Study Habit Analyzer

Welcome to the **Academic Performance & Study Habit Analyzer**! 🎓

This is a beginner-friendly project designed to help you understand how a student's daily inputs (like study hours, sleep, and attendance) relate to their academic outputs (grades). 

In data science, we often use visualizations to find patterns in data. In this notebook, you will learn how to:
1. **Prepare Data** using a library called `pandas`.
2. **Visualize Data** using a library called `matplotlib` with four different types of charts:
   * 📊 **Bar Chart:** To compare average grades across different subjects side-by-side.
   * 📈 **Line Chart:** To track your grade progress across a semester (showing improvement over time).
   * 🍕 **Pie Chart:** To see how your weekly hours are split between study, leisure, sleep, etc.
   * 🎯 **Scatter Plot:** To see the relationship (correlation) between study hours and average grades.

Let's get started! 🚀

```python

import pandas as pd


import matplotlib.pyplot as plt

print("Libraries imported successfully! Ready to analyze data.")
```

##  1. Creating Our Student Dataset

Before we can create charts, we need some data to work with. 
We will create a simple table of subjects that contains:
* **Subject:** The name of the course.
* **Study Hours (per week):** Average hours spent studying per week.
* **Attendance Rate (%):** Percentage of classes attended.
* **Average Grade (%):** The average score in that subject.

```python

data = {
    "Subject": ["Math", "Python Programming", "Science", "History", "English"],
    "Study Hours (per week)": [8, 12, 6, 4, 5],
    "Attendance Rate (%)": [95, 98, 90, 85, 88],
    "Average Grade (%)": [85, 92, 78, 70, 82]
}

# Convert the dictionary into a DataFrame (a nice table)
df = pd.DataFrame(data)

# Display the table in the notebook
df
```

## 📊 2. Bar Chart: Comparing Grades Across Subjects

A **Bar Chart** is perfect for comparing different categories side-by-side. 
Here, we will compare the average grade for each subject to see which subjects have the highest and lowest scores.

```python

plt.figure(figsize=(8, 5))

# Plot the bars. We use 'Subject' for the X-axis and 'Average Grade (%)' for the Y-axis.
# We also choose a nice color palette and outline the bars with a thin black border.
colors = ["#4F46E5", "#06B6D4", "#10B981", "#F59E0B", "#EF4444"] # A modern color palette
plt.bar(df["Subject"], df["Average Grade (%)"], color=colors, edgecolor="black", width=0.6)

# Add titles and labels so people know what the chart is showing
plt.title("Average Grades Across Subjects", fontsize=14, fontweight="bold", pad=15)
plt.xlabel("Subjects", fontsize=12, labelpad=10)
plt.ylabel("Average Grade (%)", fontsize=12, labelpad=10)

# Set the limits of Y-axis from 0 to 100 to represent percentage marks
plt.ylim(0, 100)

# Add horizontal gridlines to make it easier to read the values
plt.grid(axis="y", linestyle="--", alpha=0.5)

# Show the chart!
plt.tight_layout()
plt.show()
```

## 📈 3. Line Chart: Tracking Grade Trajectory Over the Semester

A **Line Chart** is the best choice for showing how something changes over time. 
Let's track a student's grade progress in their "Python Programming" course from the first quiz to the final exam. This will show their learning curve and improvement!

```python

milestones = ["Quiz 1", "Quiz 2", "Midterm", "Project 1", "Final Exam"]
grades = [65, 72, 80, 88, 92] # Notice how the grades improve!

# Set the figure size
plt.figure(figsize=(8, 5))


plt.plot(milestones, grades, marker="o", color="#4F46E5", linewidth=2.5, markersize=8)


plt.title("Grade Progress Trajectory Across Semester", fontsize=14, fontweight="bold", pad=15)
plt.xlabel("Semester Milestones", fontsize=12, labelpad=10)
plt.ylabel("Grade Score (%)", fontsize=12, labelpad=10)


plt.ylim(50, 100)


plt.grid(True, linestyle="--", alpha=0.5)


plt.tight_layout()
plt.show()
```

##  4. Pie Chart: Weekly Time Breakdown

A **Pie Chart** shows "parts of a whole." It is excellent for visualizing how a single resource (like the 168 hours in a week) is divided into different activities.

Let's see how a typical student splits their weekly hours:
* **Lectures:** 15 hours
* **Self-Study:** 20 hours
* **Extracurriculars:** 10 hours
* **Sleep:** 56 hours (8 hours a day)
* **Leisure/Rest:** 67 hours

```python

activities = ["Lectures", "Self-Study", "Extracurriculars", "Sleep", "Leisure"]
hours = [15, 20, 10, 56, 67]


explode = (0, 0.1, 0, 0, 0)


pie_colors = ["#3B82F6", "#10B981", "#F59E0B", "#8B5CF6", "#EC4899"]


plt.figure(figsize=(7, 7))


plt.pie(hours, explode=explode, labels=activities, colors=pie_colors, 
        autopct="%1.1f%%", shadow=True, startangle=140, 
        textprops={'fontsize': 11})


plt.title("Weekly Time Breakdown (168 Hours)", fontsize=14, fontweight="bold", pad=20)


plt.show()
```

## 🎯 5. Scatter Plot: Study Time vs. Grades (Analyzing the Correlation)

Now, let's look at the ultimate question: **Does studying more relate to getting higher grades?**

A **Scatter Plot** is the best chart to show the relationship (or correlation) between two variables: weekly study hours (our input) and average grades (our output).

```python

study_hours = df["Study Hours (per week)"]
grades_list = df["Average Grade (%)"]


plt.figure(figsize=(8, 5))


plt.scatter(study_hours, grades_list, color="#10B981", edgecolor="black", s=100, zorder=3, label="Subjects")


for i, subject in enumerate(df["Subject"]):
    plt.annotate(subject, (study_hours[i], grades_list[i]), xytext=(8, -4), textcoords='offset points', fontsize=10, fontweight="semibold")


plt.title("Study Hours vs. Average Grades", fontsize=14, fontweight="bold", pad=15)
plt.xlabel("Study Hours (per week)", fontsize=12, labelpad=10)
plt.ylabel("Average Grade (%)", fontsize=12, labelpad=10)


plt.grid(True, linestyle="--", alpha=0.5, zorder=0)


plt.xlim(2, 14)
plt.ylim(60, 100)

# Show the chart!
plt.tight_layout()
plt.show()
```
